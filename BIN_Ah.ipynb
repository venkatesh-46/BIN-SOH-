```python
#iterating each file in input directory
for f in os.listdir():
    f_name,f_ext=os.path.splitext(f) #arguments to extract filename and file extension.
    f_bin,f_Ah=f_name.split("_") #splitting filename by underscore into two components(f_bin-bin and f_Ah-Ah).
    df=pd.read_csv(f_bin+'_'+f_Ah+'.csv') #loading the file dynamically as pandas dataframe df.
    df['DateTime']=pd.to_datetime(df.DateTime) #converting Timestamp to datatime object.
    A=list(df.Charger_Output_Current) #parsing charger output current as list and assigning it to variable A.
    pd.to_numeric(A,errors='coerce')  #convert argument to numeric type.
    delta_Charger_output_current=[y-x for x,y in zip(A[:-1],A[1:])] #delta charger output current for each instance.
    delta_Charger_output_current.append(0.0) #appending 0 as last value
    delta_Charger_output_current=pd.DataFrame(delta_Charger_output_current,columns=['delta_Charger_output_current']) #converting to dataframe.
    df=pd.merge(df,delta_Charger_output_current,left_index=True,right_index=True) #merging delta COC to dataframe df.
    df.delta_Charger_output_current.value_counts() #checking value distribution and count.  
    df[df.delta_Charger_output_current<-0.1].head(1) #extracting record where the delta COC exceeds threshold.
    df[df.delta_Charger_output_current<-0.1].head(1)['DateTime'] #extracting Timestamp at threshold.
    ind=df[df.delta_Charger_output_current<-0.1].head(1)['DateTime'].index #extracting index and assigning as ind.
    df_ind=df.loc[ind,'C1':'C13'] #parsing ind to dataframe and assigning to another variable df_ind.
    df_ind.iloc[0].sort_values(ascending=False) #sorting the values in descending.
    df_ind_weakest_rank=df_ind.iloc[0].sort_values(ascending=False).rank() #assigning above values to variable and finding rank.
    df['CC']=df['Charger_Output_Current']/2 #calculating CC by dividing charger output current/2
    df['CC']=df['CC']/sec #converting AH to A.
    df['CC']=df['CC']*mins #converting AH to A.
    df['CC']=df.CC.sum() #accumulating All the values of converted Ampere and assigning as CC.
    df_voltage_old=df.iloc[0,1:14] #extracting first datapoint by slicing.
    df_voltage_old=list(df_voltage_old) #storing the values into list and assigning as df_voltage_old.
    df_voltage_old=[round(x,2) for x in df_voltage_old] #rounding the decimals in list to 2 places.
    df_voltage_new=df.iloc[-1,1:14] #extracting last datapoint by slicing.
    df_voltage_new=list(df_voltage_new) #storing the values into list and assigning as df_voltage_new.
    df_voltage_new=[round(x,2) for x in df_voltage_new] #rounding the decimals in list to 2 places.
    df_voltage_old=pd.DataFrame(df_voltage_old) #parsing df_voltage_old into dataframe.
    df_voltage_old.columns=['old_voltage'] #naming column as old_voltage.
    #mapping voltage to SOC based on Ah.
    if f_Ah=='13':
        df_soc_old=df_voltage_old['old_voltage'].map(ocv_13Ah)
    else :
        df_soc_old=df_voltage_old['old_voltage'].map(ocv_14Ah)
    df_soc_old=pd.DataFrame(df_soc_old) #mapped SOC data parsed into dataframe.
    df_soc_old.columns=['SOC_old'] #naming column as 'SOC_old'.
    df_voltage_new=pd.DataFrame(df_voltage_new) #parsing df_voltage_new into dataframe.
    df_voltage_new.columns=['new_voltage'] #naming column as new_voltage.
    #mapping voltage to SOC based on Ah.
    if f_Ah=='13':
        df_soc_new=df_voltage_new['new_voltage'].map(ocv_13Ah)
    else :
        df_soc_new=df_voltage_new['new_voltage'].map(ocv_14Ah)
    df_soc_new=pd.DataFrame(df_soc_new) #mapped SOC data parsed into dataframe.
    df_soc_new.columns=['SOC_new'] #naming column as 'SOC_new'.
    s1_old_soc_mapping=pd.merge(df_voltage_old,df_soc_old,left_index=True,right_index=True) #merging old voltage and soc old.
    s1_new_soc_mapping=pd.merge(df_voltage_new,df_soc_new,left_index=True,right_index=True) #merging new voltage and soc new.
    df_soh=pd.merge(s1_old_soc_mapping,s1_new_soc_mapping,left_index=True,right_index=True) #merging soc_old and soc_new into soh dataframe.
    df_soh['SOC_old']=df_soh['SOC_old']/100 #dividing soc_old by 100.
    df_soh['SOC_new']=df_soh['SOC_new']/100 #dividing soc_new by 100.
    df_soh['CC']=df.CC.values[0] #indexing the CC value .
    df_soh['SOC_new-SOC_old']=df_soh['SOC_new']-df_soh['SOC_old'] #calculating difference SOC_new-SOC_old.
    df_soh['Ah*delta_SOC']=int(f_Ah)*df_soh['SOC_new-SOC_old'] #Ah multiplied by delta SOC_new-SOC_old.
    df_soh['SOH']=df_soh['CC']/df_soh['Ah*delta_SOC'] #calculating SOH.
    df_soh['SOH_percent']=df_soh['SOH']*100 # SOH in percent.
    df_soh.index=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13'] #creating index same as cellname.
    df_soh['SOH_Rank']=df_soh.SOH_percent.rank() #creating rank column.
    least_SOH=df_soh.SOH_percent.min() #least SOH.
    df_weakest=df_ind.iloc[0].sort_values(ascending=False).head(3) #calculating weakest cell.
    df_weakest=pd.DataFrame(df_weakest) #parsing into dataframe.
    df_weakest=df_weakest.transpose() #transposing the dataframe.
    df_weakest.columns #extracting columns.
    w1=df_weakest.columns[0] #weakest cell index 0.
    w2=df_weakest.columns[1] #weakest cell index 1.
    w3=df_weakest.columns[2] #weakest cell index 2.
    BIN=f_bin #bin name.
    AH=f_Ah   #Ah.
    file_name=f_bin+'_'+f_Ah #file name.
    soh=least_SOH #least SOH.
    ts=str(df[df.delta_Charger_output_current<-0.1].head(1)['DateTime']) #timestamp extraction
    ts=ts[5:24] #timestamp
    vol_1=df_weakest.iloc[:,0].values
    vol_2=df_weakest.iloc[:,1].values
    vol_3=df_weakest.iloc[:,2].values
    voltage_1=vol_1[0]
    voltage_2=vol_2[0]
    voltage_3=vol_3[0]
    ind.values
    ind_1=ind[0]
    w1_v=df.loc[:ind_1,w1]
    w1_index=w1_v[w1_v==voltage_1].index[0]
    w1_ts=df.loc[w1_index,'DateTime']
    
    w2_v=df.loc[:ind_1,w2]
    w2_index=w2_v[w2_v==voltage_2].index[0]
    w2_ts=df.loc[w2_index,'DateTime']
    
    w3_v=df.loc[:ind_1,w3]
    w3_index=w3_v[w3_v==voltage_3].index[0]
    w3_ts=df.loc[w3_index,'DateTime']
    
    data={'W1_T':[w1_ts],'W2_T':[w2_ts],'W3_T':[w3_ts],'TimeStamp':[ts],'Voltage_W1':[vol_1],'Voltage_W2':[vol_2],'Voltage_W3':[vol_3],'W1':[w1],'W2':[w2],'W3':[w3],'BIN':[BIN],'FN':[file_name],'AH':[AH],'SOH':[soh]} #creating dictionary and parsing data.
    df1=pd.DataFrame(data) #converting dictionary into dataframe and assigning as df1.
    df2=df2.append(df1) #appending df1 data to new dataframe df2.
    print(df1) #output
    
```

####Computing CC parameter from Charger_Output_Current  and Calculating SOH by using formula.               SOH=CC/13*delta(soc_new-soc_old).
•Since the cells are connected in parallel the current is divided between two cells so we are dividing each instance of Charger_Output_Current  by 2.
•Converting A to Ah by dividing each instance of charger_output_current by 3600 and multiplying by 60.
•we accumulate All the A values of Charger_output_current which Equates to CC.
•SOC_OLD is Taken from initial charging time period and Mapping the corresponding cell voltages to SOC values from OCV vs SOC data.
•SOC_NEW is Taken from final charging time period and Mapping the corresponding cell voltages to SOC values from OCV vs SOC data.
•Then Calculate the difference between SOC_new and SOC_old and multiply by 13 since that much current has been given.
•Finally computing SOH from the formula SOH=CC/13*delta(soc_new-soc_old).
